# Byte Pair Encoding (BPE) Tokenizer From Scratch Simple

## Bits and Bytes

```python
text = "This is some text"
byte_ary = bytearray(text, "utf-8")
print(byte_ary)
```

```python
ids = list(byte_ary)
print(ids)
```

```python
print("Number of characters:", len(text))
print("Number of token IDs:", len(ids))
```

```python
import tiktoken
gpt2_tokenizer = tiktoken.get_encoding("gpt2")
gpt2_tokenizer.encode("This is some text")
```

```python
for i in range(300):
    decoded = gpt2_tokenizer.decode([i])
    print(f"{i}: {decoded}")
```

## A simple BPE Usage

```python
import os
import urllib.request
from BPEImplemetation import BPETokenizerSimple

if not os.path.exists("../main-chapter-code/the-verdict.txt"):
    url = ("https://raw.githubusercontent.com/rasbt/"
           "LLMs-from-scratch/main/ch02/01_main-chapter-code/"
           "the-verdict.txt")
    file_path = "../main-chapter-code/the-verdict.txt"
    urllib.request.urlretrieve(url, file_path)

with open("../main-chapter-code/the-verdict.txt", "r", encoding="utf-8") as f: # added ../01_main-chapter-code/
    text = f.read()
```

```python
tokenizer = BPETokenizerSimple()
tokenizer.train(text, vocab_size=1000, allowed_special={"<|endoftext|>"})
```

```python
print(tokenizer.vocab)
print(len(tokenizer.vocab))
```

```python
print(len(tokenizer.bpe_merges))
```

```python
input_text = "Jack embraced beauty through art and life."
token_ids = tokenizer.encode(input_text)
print(token_ids)
```

```python
print("Number of characters:", len(input_text))
print("Number of token IDs:", len(token_ids))
```

```python
print(token_ids)
```

```python
print(tokenizer.decode(token_ids))
```

```python
for token_id in token_ids:
    print(f"{token_id} -> {tokenizer.decode([token_id])}")
```

```python
tokenizer.decode(tokenizer.encode("This is some text."))
```
